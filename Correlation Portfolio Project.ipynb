```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the congiguration of the plot that we will create

#Read the dataset

#For resolving the UniCodedecode error
import chardet
file = 'G:\chrome downloads\movies.csv'
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result

df = pd.read_csv(file,encoding='ISO-8859-1')
```

```python
#Look at the data

df.head()
```

```python
# Checking if there is any missing data

for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing))
```

```python
# Checking the datatypes of the columns 

df.dtypes
```

```python
# Changing the dataypes of the columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Correcting the year relased

df['year_correct'] = df['released'].astype(str).str[:4]
df['year_correct'] = df['year_correct'].astype('int64')
```

```python
#Sorting the dataframe by the gross

df = df.sort_values(by=['gross'], inplace = False, ascending= False)
```

```python
# Display the maximum rows
pd.set_option('display.max_rows', None)
```

```python
# Let's Start looking at correlation with Budget

df.corr(method = 'spearman') #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method = 'spearman')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
df.head()
```

```python
# Let's Start looking at correlation with Company
# First we will numerize the company name to find the correlation

df_numerized = df

for col in df_numerized.columns:
    if (df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'spearman')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
